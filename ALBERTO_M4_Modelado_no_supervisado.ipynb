```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler 
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

### We have Min and Max very far away from each others. That's why we need to rescaale with MinMax Scaler:

1. Split the data into Training and test
2. MinMax scaler and we apply it only to the training data so the results won't be affected

```python
# Splitting the data
X = df.drop(columns=["Credito_tipo_1", "Credito_tipo_2", "Producto_ofrecido"])
# y = df["Producto_ofrecido"]

# Getting the columns
X_col = X.columns
```

```python
type(X)
```

```python
# We are going to use normalization since clustering doesn't make assumption about the data distribution 
# and the data is varying a lot

# Creating the object:
scaler = StandardScaler()

# Only scaling the traning so we won't alter the results on our model
X = scaler.fit_transform(X) # Normalizing and fitting (applying)

# converting it into a data frame again
X = pd.DataFrame(X, columns = X_col)
```

```python
X.describe()
```

```python
# Applying the Elbow method to see the number of clusters we need

wcss = [] # >> To store the inertia that we will be plotting

# We are going to loop through a range to see what's the best cluster. We pick an arbitrary rage to see the elbow

for i in range(1, 20): 
  
  k_means = KMeans(n_clusters=i, random_state=42)
  
  k_means.fit(X)
  wcss.append(k_means.inertia_)
  
  # End of for
```

```python
# Plotting 
fig, ax= plt.subplots(figsize=(15, 10))
ax.plot(range(1, 20), wcss, marker = "o")
ax.grid()
ax.set_title('Elbow method', fontdict={"size":16})

# X divisions
ax.set_xticks(np.arange(20))

ax.set_xlabel('Number of clusters', fontdict={"size":16})
ax.set_ylabel('Inertia', fontdict={"size":16});
```

```python
# Verifying the performance

kmedias = KMeans(n_clusters=2, random_state=42) # Creating the object
kmedias.fit(X)
labels = kmedias.labels_
# right_labels = sum(y == labels)

# Predicting the values
groups = kmedias.fit_predict(X);

# print("Resultado: %d de %d muestras correctamente etiquetadas" % (right_labels, y.size))

# print("Precision: {0:0.2f}". format(right_labels/(y.size)))
```

```python
new_cluster_df = df.copy()
new_cluster_df["Groups"] = groups
new_cluster_df.head(15)
```

```python
plt.scatter(new_cluster_df["Salario"], new_cluster_df["Limite_TC"], c = new_cluster_df["Groups"], cmap = "rainbow");
plt.title("Salario vs Limite TC", fontdict={"size":16})
plt.xlabel('Salario', fontdict={"size":16})
plt.ylabel('Limite TC', fontdict={"size":16});
```

```python
# Showing how many people from with "Credito tipo 1" are allocated on each group

np.around(new_cluster_df.groupby(["Groups"])["Credito_tipo_1"].mean()*100, 2) # Anailizar los demas typos de creditos
```

```python
# Showing how many people from with "Credito tipo 2" are allocated on each group

np.around(new_cluster_df.groupby(["Groups"])["Credito_tipo_2"].mean()*100, 2) 
```

```python
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(X)
```

```python
PCA_components = pd.DataFrame(principalComponents)
PCA_components.rename(columns={0:'Componente 1',1:'Componente 2',2:'Componente 3'},inplace=True)
df_pca = pd.concat([new_cluster_df.reset_index(),PCA_components],axis=1,sort=False).drop('index',axis=1)
df_pca
```

```python

features = range(pca.n_components_)
plt.bar(features,pca.explained_variance_ratio_.cumsum())
plt.xlabel('PCA Características')
plt.ylabel('% Varianza explicada')
```

```python
cols = ['Salario','Limite_TC','Tiempo_cliente']
componentes = pd.DataFrame(pca.components_)
componentes.insert(0,'vars',cols)
componentes.set_index('vars',inplace=True)
componentes.rename(columns={i:f"Componente {i+1}" for i in range(len(cols))},inplace=True)
componentes
```

```python
plt.scatter(df_pca["Componente 1"], df_pca["Componente 2"], c = df_pca["Groups"], cmap = "rainbow");
plt.title("Componente 1 vs Componente 2", fontdict={"size":16})
plt.xlabel('Componente 1', fontdict={"size":16})
plt.ylabel('Componente 2', fontdict={"size":16});
```
