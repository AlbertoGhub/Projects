This is a regression problem. WE are going to pick California Housing dataset.

```python
# Libraries:

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score
```

```python
# This time, we aren't going to download the dataset. Instead, we are going to 
# import it directly from the sklearn library
# California Housing dataset - https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_california_housing.html#sklearn.datasets.fetch_california_housing
# https://scikit-learn.org/stable/datasets/real_world.html#california-housing-dataset

from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
housing
```

```python
# The data is not shown in a correct format, so we will convert it into a dataset

housing_df = pd.DataFrame(housing["data"], columns=housing["feature_names"])
housing_df
```

Now we have the data, now, to have the dataframe completed, we will add the target column so we can see it properly

```python
housing_df["MedHouseVal"] = housing["target"]
housing_df.head(15)
```

```python
housing_df.dtypes
```

# Meaning of the values:

  + MedInc: median income in block group

  + HouseAge: median house age in block group

  + AveRooms: average number of rooms per household

  + AveBedrms: average number of bedrooms per household

  + Population: block group population

  + AveOccup: average number of household members

  + Latitude: block group latitude

  + Longitude: block group longitude


This dataset was derived from the 1990 U.S. census, using one row per census block group. A block group is the smallest geographical unit for which the U.S. Census Bureau publishes sample data (a block group typically has a population of 600 to 3,000 people).

```python
# After the data is formatted the way we want, now we are constructing the model
# Let's do it

# Importing the model (algorithm)/estimator

# setup random seed
np.random.seed(42)

# Create the data
X = housing_df.drop("MedHouseVal", axis = 1)
y = housing_df["MedHouseVal"] # the median house value for California districts, expressed in hundreds of thousands of dollars ($100,000).

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

# Let's create the model and fit it into the training set
model = Ridge()
model.fit(X_train, y_train)

# Check the accurecy (score) of the model
print(f"The precision is for this model is: {np.around(model.score(X_test, y_test)*100, 2)}%")
```

# The thing with Machine learning is that you have to try methods until you have a better performance

```python
# RandomForestClassifier
model = RandomForestRegressor()

# Creating the model
model.fit(X_train, y_train)
```

```python
print(f"The score for random forest is {np.around(model.score(X_test, y_test), 4)*100}%")
```

# Another way of calculating the accuracy of a regression model.

```python
y_preds = model.predict(X_test)
```

```python
print(f"The accuracy with the R2 score is:\n\n{np.around(r2_score(y_test, y_preds)*100, 4)}%")
```

# We can keep on trying another metrics to evaluate our model or simply look for another one that provides us a better performance
